# Сохраняет вертикально

```python
from PIL import Image, ImageDraw
import random

def add_pattern_to_image(image_path, pattern_path, output_path, txt_file_path):
    # Открываем изображение
    with Image.open(image_path) as img:
        # Открываем паттерн
        with Image.open(pattern_path) as pattern:
            # Масштабируем паттерн до нужных размеров
            pattern = pattern.resize((183, 48))

            # Генерируем случайные координаты для размещения паттерна на первом изображении
            x1 = random.randint(0, img.width - pattern.width)
            y1 = random.randint(0, img.height - pattern.height)

            # Проверяем, есть ли альфа-канал в паттерне
            if pattern.mode == 'RGBA':
                # Если альфа-канал есть, используем его для маскирования
                img.paste(pattern, (x1, y1), pattern)
            else:
                # Если альфа-канала нет, создаем новую картинку с альфа-каналом
                new_pattern = Image.new("RGBA", pattern.size, (255, 255, 255, 255))
                new_pattern.paste(pattern, (0, 0))
                img.paste(new_pattern, (x1, y1), new_pattern)

            # Получаем точки (углы) bbox паттерна на первом изображении
            points1 = [(x1, y1), (x1 + pattern.width, y1), (x1 + pattern.width, y1 + pattern.height), (x1, y1 + pattern.height)]
            print("Информация о точках и координатах:")
            print(f"  Точки (углы) паттерна на изображении: {points1}")

            # Создаем второе изображение (белое)
            white_img = Image.new("RGBA", img.size, (255, 255, 255, 255))

            # Объединяем изображения вертикально
            combined_img = Image.new("RGBA", (img.width, img.height + white_img.height), (255, 255, 255, 255))
            combined_img.paste(img, (0, 0))
            combined_img.paste(white_img, (0, img.height))

            # Размещаем паттерн по середине второго изображения
            x2 = (combined_img.width - pattern.width) // 2
            y2 = img.height + (white_img.height - pattern.height) // 2

            # Проверяем, есть ли альфа-канал в паттерне
            if pattern.mode == 'RGBA':
                # Если альфа-канал есть, используем его для маскирования
                combined_img.paste(pattern, (x2, y2), pattern)
            else:
                # Если альфа-канала нет, создаем новую картинку с альфа-каналом
                new_pattern2 = Image.new("RGBA", pattern.size, (255, 255, 255, 255))
                new_pattern2.paste(pattern, (0, 0))
                combined_img.paste(new_pattern2, (x2, y2), new_pattern2)

            # Получаем точки (углы) bbox паттерна на втором изображении
            points2 = [(x2, y2), (x2 + pattern.width, y2), (x2 + pattern.width, y2 + pattern.height), (x2, y2 + pattern.height)]
            print(f"  Точки (углы) паттерна на белом изображении: {points2}")

            # Сохраняем результат объединенного изображения в формате PNG
            combined_img.save(output_path.replace('.jpg', '_combined.png'))

            # Создаем список строк в нужном формате для объединенного изображения
            result_list_combined = [f"{coord[0] / combined_img.width} {coord[1] / combined_img.height}" for coord in points1 + points2]

            # Объединяем список в одну строку для объединенного изображения
            coordinates_string_combined = " ".join(result_list_combined)

            # Записываем координаты в текстовый файл для объединенного изображения
            with open(txt_file_path.replace('.txt', '_combined.txt'), "w") as txt_file_combined:
                txt_file_combined.write(f"0 {coordinates_string_combined}")
                print(f"  Координаты (нормализованные) объединенные: {coordinates_string_combined}")

# Пути к файлам
pattern_path = r"D:\DataScience\Work\advertising_blurring\dataset\pattern\winline_1.1.png"
image_path = r"D:\DataScience\Work\advertising_blurring\dataset\frames\v1\frame_0.jpg"
output_path = r"D:\DataScience\Work\advertising_blurring\notebooks\tmp.jpg"
txt_file_path = r"D:\DataScience\Work\advertising_blurring\notebooks\coordinates.txt"

# Добавляем паттерн на изображение и сохраняем результат для первого и второго изображений
add_pattern_to_image(image_path, pattern_path, output_path, txt_file_path)

# Выводим пути к созданным файлам
# Выводим информацию о сохраненных файлах
print("")
print("Сохраненные файлы:")
print(f"  Координаты: {txt_file_path}")
print(f"  Первое изображение: {output_path}")
print(f"  Второе изображение (объединенное): {output_path.replace('.jpg', '_combined.png')}")
print(f"  Координаты второго изображения (объединенного): {txt_file_path.replace('.txt', '_combined.txt')}")
```

```python
from pathlib import Path
from PIL import Image
import random
import os
import glob

def add_pattern_to_image(image_path, pattern_path, output_folder, txt_folder, image_folder):
    """
    Добавляет паттерн на изображение и сохраняет результат.

    Parameters:
    - image_path (str): Путь к изображению.
    - pattern_path (str): Путь к паттерну.
    - output_folder (str): Папка для сохранения изображений.
    - txt_folder (str): Папка для сохранения координат в текстовом файле.
    - image_folder (str): Папка с изображениями.

    Returns:
    None
    """
    image_folder_path = Path(image_path).parent.relative_to(Path(image_folder))
    output_subfolder = Path(output_folder) / image_folder_path
    txt_subfolder = Path(txt_folder) / image_folder_path

    with Image.open(image_path) as img:
        with Image.open(pattern_path) as pattern:
            pattern = pattern.resize((183, 48))
            x1 = random.randint(0, img.width - pattern.width)
            y1 = random.randint(0, img.height - pattern.height)

            if pattern.mode == 'RGBA':
                img.paste(pattern, (x1, y1), pattern)
            else:
                new_pattern = Image.new("RGBA", pattern.size, (255, 255, 255, 255))
                new_pattern.paste(pattern, (0, 0))
                img.paste(new_pattern, (x1, y1), new_pattern)

            points1 = [(x1, y1, x1 + pattern.width, y1 + pattern.height)]

            white_img = Image.new("RGBA", img.size, (255, 255, 255, 255))

            combined_img = Image.new("RGBA", (img.width, img.height + white_img.height), (255, 255, 255, 255))
            combined_img.paste(img, (0, 0))
            combined_img.paste(white_img, (0, img.height))

            x2 = (combined_img.width - pattern.width) // 2
            y2 = img.height + (white_img.height - pattern.height) // 2

            if pattern.mode == 'RGBA':
                combined_img.paste(pattern, (x2, y2), pattern)
            else:
                new_pattern2 = Image.new("RGBA", pattern.size, (255, 255, 255, 255))
                new_pattern2.paste(pattern, (0, 0))
                combined_img.paste(new_pattern2, (x2, y2), new_pattern2)

            points2 = [(x2, y2, x2 + pattern.width, y2 + pattern.height)]

            # Выводим информацию о точках (углах) паттерна
            print(f"  Точки (углы) паттерна на белом изображении: {points1}")
            print(f"  Точки (углы) паттерна на белом изображении: {points2}")

            # Создаем поддиректории, если они не существуют
            output_subfolder.mkdir(parents=True, exist_ok=True)
            txt_subfolder.mkdir(parents=True, exist_ok=True)

            # Путь для сохранения объединенного изображения
            output_path = output_subfolder / f"{Path(image_path).stem}_combined.png"
            combined_img.save(output_path)

            # Путь для сохранения координат объединенного изображения
            txt_file_path_combined = txt_subfolder / f"{Path(image_path).stem}_combined.txt"
            with open(txt_file_path_combined, "w") as txt_file_combined:
                result_list_combined = [f"0 {coord[0] / combined_img.width} {coord[1] / combined_img.height} {coord[2] / combined_img.width} {coord[3] / combined_img.height}" for coord in points1 + points2]
                coordinates_string_combined = "\n".join(result_list_combined)
                txt_file_combined.write(coordinates_string_combined)

def process_images_in_folder(image_folder, pattern_folder, output_folder, txt_folder):
    image_paths = glob.glob(os.path.join(image_folder, '**', '*.png'), recursive=True)

    if not image_paths:
        print(f"Нет изображений для обработки в {image_folder}.")
    else:
        patterns = glob.glob(os.path.join(pattern_folder, '*.png'))

        if not patterns:
            print("Нет паттернов для обработки.")
        else:
            for image_path in image_paths:
                # Choose a pattern for each image
                pattern_path = random.choice(patterns)

                add_pattern_to_image(image_path, pattern_path, output_folder, txt_folder, image_folder)

                print("")
                print("Сохраненные файлы:")
                print(f"  Координаты: {txt_folder}")
                print(f"  Первое изображение: {output_folder}")
                print(f"  Второе изображение (объединенное): {output_folder.replace('.png', '_combined.png')}")
                print(f"  Координаты второго изображения (объединенного): {txt_folder.replace('.txt', '_combined.txt')}")

def main():
    pattern_folder = r"D:\DataScience\Work\advertising_blurring\dataset\pattern"
    input_image_folder = r"D:\DataScience\Work\advertising_blurring\notebooks\new_size_v2"
    output_folder = r"D:\DataScience\Work\advertising_blurring\notebooks\dataset_640_size"
    txt_folder = output_folder

    process_images_in_folder(input_image_folder, pattern_folder, output_folder, txt_folder)

if __name__ == "__main__":
    main()
```
